## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Setting a variable for a number of unique items
total_unique_items = clients_data["Item Name"].nunique()

#Setting a variable for an average purchase price
average_price = clients_data["Price"].mean()

#Setting a variable for total number of purchases
total_purchases = clients_data.shape[0]

#Setting a variable for a total revenue
total_revenue = clients_data["Price"].sum()

#Creating a dataframe with results
purchasing_analysis = pd.DataFrame(
    {"Number of Unique Items": [total_unique_items],
     "Average Purchase Price": [average_price],
     "Total Number of Purchases": [total_purchases],
     "Total Revenue": [total_revenue]
     }
)

#Displaing the result
purchasing_analysis
```

```python
#Formatting financial information in the dataframe
purchasing_analysis["Average Purchase Price"] = purchasing_analysis["Average Purchase Price"].map(
    "${:,.2f}".format)
purchasing_analysis["Total Revenue"] = purchasing_analysis["Total Revenue"].map(
    "${:,.2f}".format)

purchasing_analysis
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Creating age bins 
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

# Creating the names for bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
#Creating a new column in existing dataframe
#with age backets data
clients_data["Age Group"] = pd.cut(clients_data["Age"], bins, labels=group_labels, include_lowest = False)
#Displaying data
clients_data
```

```python
#Creating a new dataframe with values
#for using in this and the next task
age_demographic_total = pd.DataFrame(clients_data.groupby("Age Group").nunique())
age_demographic_total
```

```python
#Creating a dataframe with age demographic results
age_demographic = pd.DataFrame(age_demographic_total.loc[:, "SN"])
age_demographic
```

```python
#Renaming a column
age_demographic.rename(columns={"SN":"Count"}, inplace = True)
age_demographic
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Creating a new dataframe for farther groupby
items = clients_data.loc[:, ["Item ID", "Item Name", "Price"]]
items.head()
```

```python
#Creating a new dataframe according to the task
popular_items = pd.DataFrame(items.groupby(["Item ID", "Item Name"])["Item ID"].size())
popular_items
```

```python
#renaming the column
popular_items = popular_items.rename(columns={"Item ID":"Purchase Count"})
popular_items
```

```python
#Counting an average price and adding it to the dataframe
popular_items["Item Price"] = items.groupby(["Item ID", "Item Name"])["Price"].mean()
popular_items
```

```python
#Counting a total value of each item and adding it to the dataframe
popular_items["Total Purchase Value"] = items.groupby(["Item ID", "Item Name"])["Price"].sum()
popular_items
```

```python
#Sorting the data frame in order to see most popular items
most_popular_items = popular_items.sort_values("Purchase Count", ascending=False)
most_popular_items.head()
```

```python
#Checking data types in the data frame
most_popular_items.dtypes
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order

```python
#Creating a new dataframe with sorted values
most_profitable_items = most_popular_items.sort_values(["Total Purchase Value"], ascending=False)
most_profitable_items.head()
```
